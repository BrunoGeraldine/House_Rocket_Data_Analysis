## Imports

```python
import numpy as np
import pandas as pd
```

## Load DataSet

```python
data = pd.read_csv( 'datasets/kc_house_data.csv' )
data.head()
```

### 1. Create a new column called: “house_age”
    • If the value of the column “date” is greater than 2014-01-01 => ‘new_house’
    • If the value of the column “date” is less than 2014-01-01 => ‘old_house’

```python
# First, shift to object dat for date type.
data['date'] = pd.to_datetime(data['date'], format='%Y-%m-%d')
# Creating new cloumn "house_age"and attribuiting values to new column whit .apply comand.
data['house_age'] = data['date'].apply(lambda x: 'new_house' if x > pd.to_datetime('2014-01-01', format='%Y-%m-%d')  else 'old_house')
data.head(3)
```

### 2. Create a new column "dormitory_type@
    • If the value of the “bedrooms” column is equal to 1 => ‘studio’
    • If the value of the “bedrooms” column is equal to 2 => ‘apartment’
    • If the “bedrooms” column value is greater than 2 => ‘house’

```python
# Creating new cloumn "dormitory_type" and attribuiting values to new column whit .apply comand.
data['dormitory_type'] = data['bedrooms'].apply(lambda x:   'studio' if x == 1 else
                                                            'apartment' if x == 2 else 
                                                            'house' if x > 2 else 'NA')
data.head(3)
```

### 3. Create a new column called: “condition_type”
    • If the value of the column "condition" is less than or equal to 2 => ‘bad’
    • If the value of the column "condition" is equal to 3 or 4 => ‘regular’
    • If the value of the column “condition” is equal to 5 => ‘good’

```python
data['condition_type'] = data['condition'].apply(lambda x:  'bad' if x <= 2 else
                                                            'regular' if (x == 3) | (x == 4) else
                                                            'good')
data.head(3)
```

### 4. Change the TYPE Column "condition" to STRING

```python
data['condition'] = data['condition'].astype(str) 
data.dtypes
```

### 5. Delete the columns: “sqft_living15” and “sqft_lot15”

```python
# Deleting the columns and renomed the dataframe
df = data.drop(columns=['sqft_living15', 'sqft_lot15'])
df.head(3)
```

### 6. Change the TYPE Column “yr_build” to DATE

```python
df['yr_built'] = pd.to_datetime(df['yr_built'], format='%Y')
df['yr_built']
```

### 7. Change the TYPE Column “yr_renovated” to DATE


```python
# Shift the type of "yr_renovated" column
# Select data that has a value of 0 and shifts to 1900 years.
df['yr_renovated'] = df['yr_renovated'].apply(lambda x: pd.to_datetime('1900-01-01', format='%Y-%m-%d') if x == 0  else 
                                                        pd.to_datetime(x, format='%Y'))
#df.dtypes
```

### 8. What is the earliest date of construction of a property?


```python
# Ploting only year date whit ".year"
min_year_built = df['yr_built'].min().year
print('The earliest date of construction is: {}'. format(min_year_built))
```

### 9. What is the earliest date for renovation of a property?


```python
# Select data where some renewal has taken place, and take minimum value.
min_year_renovated = df.loc[df['yr_renovated'] > pd.to_datetime('1900-01-01', format='%Y-%m-%d'), 'yr_renovated'].min()
print('The earliest date of renovation is: {}'. format((min_year_renovated).year))
```

### 10. How many properties have 2 floors?


```python
house_2_floors = df.loc[df['floors'] == 2, 'id'].size
print('There are {} houses whit 2 floors'.format(house_2_floors))
```

### 11. How many properties are with the condition equal to “regular”?


```python
print('There are {} house at "regular" condition'.format(df.loc[df['condition_type'] == 'regular', 'id'].size))
```

### 12. How many properties are in “bad” condition and have “water views”?


```python
water_bad = df.loc[(df['condition_type'] == 'bad') & (df['waterfront'] == 1) , 'id'].size
print('There are {} houses whit waterfront view and "bad" contitions'.format(water_bad))
```

### 13. How many properties are “good” and “new_house”?


```python
good_new_house = df.loc[(df['condition_type'] == 'good') & (df['house_age'] == 'new_house') , 'id'].size
print('There are {} houses at "good" condition and "new house" age'.format(good_new_house))
```

### 14. What is the value of the most expensive property of the “studio” type?


```python
max_studio = df.loc[df['dormitory_type'] == 'studio', 'price'].max()
print('the most expensive "studio" type property costs U${}'.format(max_studio))
```

### 15. How many “apartment” properties were renovated in 2015?


```python
renovated_2015 = df.loc[df['yr_renovated'] == pd.to_datetime('2015-01-01'), 'id'].size
print('There are {} houses renovated in 2015'.format(renovated_2015))
```

### 16. What is the largest number of rooms that a property of the “house” type has?


```python
max_bedrooms = df.loc[df['dormitory_type'] == 'house', 'bedrooms'].max()
print('The largest number of rooms that a property of the “house” type has is {} bedrooms'.format(max_bedrooms))
```

### 17. How many “new_house” properties were renovated in 2014?


```python
#renovated_2015 = df.loc[df['yr_renovated'] == pd.to_datetime('2015-01-01'), 'id'].size
renovated_2014 = df.loc[(df['house_age'] == 'new_house' )& 
       (df['yr_renovated'] == pd.to_datetime('2014-01-01', format='%Y-%m-%d')), 'id'].size 
print('In 2014, {} houses in "new house" condition were renovated'.format(renovated_2014))
```
